```python
totalMaleCount = 0
totalFemaleCount =0
totalNoGender =0

totalFamilies = len(df)
print('totalFamilies', totalFamilies)

for family in range(0,totalFamilies-1):
    maleCount = 0
    femaleCount =0
    noGender =0
    def updateGenderCount(person):
        global maleCount, femaleCount, noGender
        gender = person.iloc[0,2]
        if gender == 'पुरुष':
            maleCount = maleCount+1
        elif gender == 'स्त्री':
            femaleCount = femaleCount+1
        else:
            noGender = noGender+1

    FamilyPersons = [range(1,15), range(34,48), range(48,62), range(62,76), range(76,90), range(90,104), range(104,118), range(118,132), range(132,146), range(146,160), range(160,174), range(174,188)]

    for i in FamilyPersons:
        person = df.iloc[[family],i]
        anotherPerson = person.iloc[0,13]
        updateGenderCount(person)
        if anotherPerson == 'नाही':
            break
    totalMaleCount += maleCount
    totalFemaleCount += femaleCount
    totalNoGender += noGender
print('totalMaleCount', totalMaleCount)
print('totalFemaleCount', totalFemaleCount)
print('totalNoGender', totalFemaleCount)
totalPopulation = totalMaleCount + totalFemaleCount + totalNoGender
print('totalPopulation', totalPopulation)
```
