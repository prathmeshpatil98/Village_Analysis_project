```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

def collectallbirthdatedata():
    all_birthdate_data = []
    
    index_positions = [2, 35, 49, 63, 77, 91, 105, 119, 133, 147, 161, 175]
    
    for index in index_positions:
        birthdate_values = df.iloc[:, index].tolist()
        
        all_birthdate_data.append(birthdate_values)
    
    return all_birthdate_data

all_birthdate_data = collectallbirthdatedata()
```

```python


def calculate_age(birthdate):
    try:
        today = datetime.today()
        birthdate = datetime.strptime(birthdate, '%m/%d/%Y')
        age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
        return age
    except ValueError:
        return None

ages = []

for birthdate in all_birthdate_data[0][1:]:
    age = calculate_age(birthdate)
    if age is not None:
        ages.append(age)

print(ages)
```

```python

school_ranges = {
    'preschool': (0, 4),
    'Nursery': (4, 5), 
    'Junior KG': (5, 6),
    'Senior KG ': (6, 7),
    'STD-1 to STD-7': (8, 14),
    'STD8 to STD-10': (13, 17),
    'COllege (11,12)': (16, 19),
    'Bachelors': (19, 22),
    'Middle adulthood.': (22, 51),
    'adulthood': (50, 101),
    'Century cross people': (100,201),
}

age_counts = {group: 0 for group in school_ranges}

for sublist in all_birthdate_data:
    for birthdate in sublist[1:]:
        if birthdate:
            birthdate_obj = datetime.strptime(birthdate, '%m/%d/%Y')
            age = datetime.today().year - birthdate_obj.year - ((datetime.today().month, datetime.today().day) < (birthdate_obj.month, birthdate_obj.day))
            
            for group, (lower, upper) in school_ranges.items():
                if lower <= age < upper:
                    age_counts[group] += 1

total_people = sum(age_counts.values())

for group, count in age_counts.items():
    print("For the {}: {} people".format(group, count))

print("Total people across all age groups: {}".format(total_people))   
```
