```python

from datetime import datetime

data = {
    '6/1/1972': 7,
    '6/1/1973': 6,
    '6/1/1968': 6,
    'UNKNOWN': 6,
    '6/1/1970': 5,
    '6/1/1957': 5,
    '6/1/1960': 5,
    '6/1/1966': 5,
    '6/1/1978': 4,
    '6/1/1979': 4,
    '6/1/1963': 4,
    '6/1/1956': 4,
    '6/1/1958': 3,
    '6/1/1975': 3,
    '6/1/1974': 3,
    '6/1/1959': 3,
    '6/1/1953': 3,
    '6/1/1951': 3,
    '6/1/1962': 3,
    '5/27/1978': 2,
    '6/1/1943': 2,
    '6/1/1947': 2,
    '1/1/1963': 2,
    '6/1/1948': 2,
    '1/1/1954': 2
}

def calculate_age(birthdate):
    if birthdate == 'UNKNOWN':
        return None
    
    today = datetime.today()
    birthdate_obj = datetime.strptime(birthdate, '%m/%d/%Y')
    age = today.year - birthdate_obj.year - ((today.month, today.day) < (birthdate_obj.month, birthdate_obj.day))
    return age

ages = {}
for birthdate, age in data.items():
    calculated_age = calculate_age(birthdate)
    ages[birthdate] = {
        'given_age': age,
        'calculated_age': calculated_age
    }

for birthdate, age_data in ages.items():
    print(f"Birthdate: {birthdate}, Given Age: {age_data['given_age']}, Calculated Age: {age_data['calculated_age']}")
```

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta

data =  {
    '6/1/1972': 7,
    '6/1/1973': 6,
    '6/1/1968': 6,
    'UNKNOWN': 6,
    '6/1/1970': 5,
    '6/1/1957': 5,
    '6/1/1960': 5,
    '6/1/1966': 5,
    '6/1/1978': 4,
    '6/1/1979': 4,
    '6/1/1963': 4,
    '6/1/1956': 4,
    '6/1/1958': 3,
    '6/1/1975': 3,
    '6/1/1974': 3,
    '6/1/1959': 3,
    '6/1/1953': 3,
    '6/1/1951': 3,
    '6/1/1962': 3,
    '5/27/1978': 2,
    '6/1/1943': 2,
    '6/1/1947': 2,
    '1/1/1963': 2,
    '6/1/1948': 2,
    '1/1/1954': 2
}

def calculate_age(birthdate):
    if birthdate == 'UNKNOWN':
        return None
    
    today = datetime.today()
    birthdate_obj = datetime.strptime(birthdate, '%m/%d/%Y')
    
    delta = relativedelta(today, birthdate_obj)
    age = {
        'years': delta.years,
        'months': delta.months,
        'days': delta.days
    }
    return age

ages = {}
for birthdate, age in data.items():
    calculated_age = calculate_age(birthdate)
    ages[birthdate] = {
        'given_age': age,
        'calculated_age': calculated_age
    }

for birthdate, age_data in ages.items():
    print(f"Birthdate: {birthdate}, Given Age: {age_data['given_age']}, Calculated Age: {age_data['calculated_age']}")
```
